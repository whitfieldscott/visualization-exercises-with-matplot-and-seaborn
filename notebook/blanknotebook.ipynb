```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as pyplot
import seaborn as sns
```

```python
iris_df = sns.load_dataset('iris')
iris_df.head(100)
```

```python
# Group by species
grouped = iris_df.groupby("species")

# Mean sizes
means = grouped.mean()

# Variation (standard deviation)
variation = grouped.std()

print("Average sizes in cm (biggest on average):\n", means)
print("\nVariation in sizes in cm:\n", variation)
```

```python
#variable that creates a list of details we want to categorize and get answers to
features = ["sepal_length", "sepal_width", "petal_length", "petal_width"]

# for loop going through features to find info
for feature in features:
    biggest = means[feature].idxmax()
    most_variable = variation[feature].idxmax()
    print(f"{feature.capitalize()}:")
    print(f"  Biggest on average → {biggest}")
    print(f"  Most variation     → {most_variable}\n")
```

```python
import matplotlib.pyplot as plt

# Sample data
flowers = ['Setosa', 'Versicolor', 'Virginica']
avg_petal_length = [1.46, 4.26, 5.55]  # average petal length from Iris dataset

# Create bar chart
plt.figure(figsize=(6,4))
plt.bar(flowers, avg_petal_length, color='skyblue')

# Add titles and labels
plt.title("Average Petal Length by Species")
plt.xlabel("Species")
plt.ylabel("Petal Length (cm)")
plt.grid()

# Show plot
plt.show()
```
